## Предобработка данных

Определим и изучим пропущенные значения. Приведем данные к нужным типам.

### Пропуски значений и типы данных
Посчитаем количество пропущенных значений в таблице:

```python
df.isna().sum() # подсчёт пропусков
```

Итак, мы видим множество столбцов с пропущенными значениями типа NaN (специальное значение типа float, которое используется если результат вычисления не может быть представлен как конкретное число, либо неизвестен). 

Пропуски в данных появляются разными путями. Например, пользователь не указал в объявлении какие-нибудь сведения, или автоматизированная система сбора информации дала сбой. Иногда пропуски оставляют умышленно, рассчитывая на автозаполнение спецсимволами.
На практике было бы правильно установить причину пропусков и восстановить данные. Такой возможности нет, поэтому действия следующие.

Для некоторых пропущенных значений можно предположить логичную замену. Например, если человек не указал число балконов — скорее всего, их нет. Такие пропуски правильно заменить на 0. Для других типов данных нет подходящего значения на замену. В этом случае правильно оставить эти значения пустыми. 

```python
df['balcony'] = df['balcony'].fillna(value='0') #сделаем логичную замену, в столбце 'balcony' пропуки отметим как их отсутствие
```

```python
df['balcony'].value_counts() # просмотрим уникальные значения столбца, и заодно оценим их холичество
```

В столбце `balcony` мы можем увидеть несоответсвие форматов, изменением типов данных мы займемся после обработки всех пропусков.

Рассмотрим столбец `ceiling_height` - высота потолков в метрах. Столбец содержит 9195 пропущенных значений, а так же изучив описание таблицы, можно увидеть аномальные значения. Минимальная высота потолков - 1 метр, максимальная - 100 метров. Можно предположить, что эти ошибки возникли из-за человеческого фактора, при заполнении данных. Рассмотрм количество аномальных значений, и внесем исправления. Заменим отсутствующие значения на медианное значение в столбце, в зависимости от населенного пункта, в котором распологается объект. Можно предположить, что это приемлемое решение, т.к. в населенном пункте может преобладать определенный тип застройки.   

```python
df['ceiling_height'].hist(range = (2, 5), bins=100, figsize=(10,5)) # построим гистограмму распределения количества помещений
plt.show()                                                          # с определенной высотой потолков
```

Исходя из графика, можно сделать вывод, что большая часть помещений имеют высоту от 2,5 - 4 метров. Значения не входящие в этот диапазон можно считать аномальными. Поэтому мы заменим их на правдоподобные - значения менее 2.5 метров на наиболее частую величину 2.5 метров, а значения больше 4 на значения медианные для населенного пункта.

```python
df.sort_values(by='ceiling_height', ascending=False).head()#просмотрим строки таблицы с аномальными значениями высоты потолков
```

Очевидно, что аномальные значения высоты потолка возникли по причине человеческого фактора, эти ошибки возникли при заполнении параметров объекта.

```python
df['ceiling_height'].value_counts().head(15) # рассмотрим наиболее часто встречающееся значение высоты, 
                                             # чтобы точно определить границы диапазона
```

```python
df.loc[df['ceiling_height'] < 2.5, 'ceiling_height'] = 2.5 # заменим все значения столбца меньше '2.5' на '2.5'
```

```python
df.loc[df['ceiling_height'] > 4, 'ceiling_height'] = df.groupby('locality_name')['ceiling_height'].transform("median")
```

```python
df['ceiling_height'].value_counts().head(10) # оценим изменения
```

Таким же способом обработаем пропуски. Заменим отсутствующие значения на медианные, в зависимости от населенного пункта, где расположены помещения.

```python
df['ceiling_height'].median() # выведем на экран медианное значение
```

Медиана соответствует стандартной высоте потолка, поэтому можно предположить, что там где имеются пропуски - высота потолка стандартная.

```python
df['ceiling_height'] = (df['ceiling_height'].                                      # Заменим пропуски на медианные значения
                        fillna(df.groupby('locality_name')['ceiling_height'].      # в зависимости от гороада
                        transform("median"))
                       )
```

```python
df['ceiling_height'].value_counts().head()
```

```python
df['ceiling_height'].isna().sum() # проверим качество обработки пропусков
```

```python
df[df['ceiling_height'].isna()].head(5) # просмотрим строки с необработанными пропусками
```

Не все пропуски заменились на медианные значения, т.к. встречаются обявления с населенным пунктом, в котором присутствует только один объект на продажу, поэтому оставшиеся пропуски заменим на медианное значение, расчитанное по всему датафрейму.

```python
ceiling_height_median = df['ceiling_height'].median() # сохраним медиану в переменную
```

```python
df['ceiling_height'] = df['ceiling_height'].fillna(ceiling_height_median) # заменим пропущенные значения на медианные
```

```python
df['ceiling_height'].value_counts().head(10)
```

```python
df['ceiling_height'].isna().sum() # проверим качество обработки пропусков
```

Рассмотрим столбец `is_apartment`. Количество пропусков в столбце 20924. Логично предположить, что помещения без указания сведений в данной графе, не являются апартаментами. Поэтому их можно заменить на булево значение 'False'.

```python
df['is_apartment'] = df['is_apartment'].fillna(False) # заменим все пропущеные значения на булево значение False
```

В столбце `locality_name` имеется небольшой процент пропусков - отсутствие сведений о местоположении объекта недвижимости, такие объявления не представляют для нас интереса, и их можно удалить из датафрэйма. Расчитаем процент таких объявлений.

```python
pass_locality_name = df['locality_name'].isna().sum()/len(df['locality_name']) # доля пропусков в общем числе объявлений
```

```python
df = df.dropna(subset=['locality_name']) # удалил строки с пропущеным значением столбца `locality_name`
```

Прежде чем обработать пропуски в оставшихся столбцах, приведем данные столбцов в числовой формат.

```python
dtype_to_replace = ['floors_total','balcony','parks_around_3000',
                    'ponds_around_3000','days_exposition',
                    'living_area','kitchen_area','airports_nearest',
                    'city_сenters_nearest','parks_nearest','ponds_nearest',
                    'days_exposition'
                   ]

for dtype in dtype_to_replace:
        df[dtype] = pd.to_numeric(df[dtype], errors='ignore')
```

```python
df.info()
```

В столбцах `kitchen_area`, `living_area` пропущенные значения заменим путем умножения общей площади жилого помещения на средний показатель отношения площади кухни/жилой площади к общей площади по всем объектам.

```python
df_living_area = df[df['living_area'] != np.nan] # создадим отфильтрованную таблицу без пропущеных значений
```

```python
average_ratio_living_area = (df_living_area['living_area']/df_living_area['total_area']).mean() # найдем среднюю величину 
                                                                                                # отношения
                                                                                                # площади жилой к общей площади
```

```python
print(average_ratio_living_area)
```

```python
df_kitchen_area = df[df['kitchen_area'] != np.nan] # создадим отфильтрованную таблицу без пропущеных значений
```

```python
average_ratio_kitchen_area = (df_kitchen_area['kitchen_area']/df_kitchen_area['total_area']).mean() # найдем среднюю величину 
                                                                                                    # отношения
                                                                                                # площади кухни к общей площади
```

```python
print(average_ratio_kitchen_area)
```

```python
living_area_to_replace = df['total_area']*average_ratio_living_area     # умножим общую площадь на расчитаный коэффициент
```

```python
df['living_area'] = df['living_area'].fillna(0)                         # заменим пропущенные значения на 0
```

```python
df.loc[df['living_area'] == 0, 'living_area'] = living_area_to_replace  # заменим нули на скорректированные значения
```

Присвоим помещениям категории в соответствии с этажом, на котором они располагаются. Обозначим помещения на первом и на последнем этаже, остальным присвоим категорию 'Другой'. Не забудем так же, что у нас имеется часть объявлений, где не указана этажность здания, поэтому нельзя определить, расположено ли помещение на последнем этаже.

```python
def floors_group(row):                      #Возвращает категорию этажа на котором расположено помещение, используя правила:   
        if row['floor'] == 1:                          # 'Первый' если помещение на 1-м этаже
            return 'Первый'                            # 'Последний' если помещение на этаже, соответствующем общей этажности
        if row['floor'] == row['floors_total']:        # 'Другой' во всех остальных случаях
            return 'Последний'                         
        return 'Другой'
```

#### *Влияние параметров на стоимость квартиры*

Изучим какие факторы больше всего влияют на стоимость квартиры. Изучим, зависит ли цена от площади, числа комнат, удалённости от центра. Изучим зависимость цены от того, на каком этаже расположена квартира: первом, последнем или другом. Также изучим зависимость от даты размещения: дня недели, месяца и года.

Каждый из этих факторов по-своему влияет на конечную стоимость квартиры. Узнать, как именно и насколько сильно — это аналитическая задача, которая решается изучением совместного распределение величин.
Взаимосвязь двух или нескольких величин называется «корреляция».

```python
columns_to_corr = ['total_area',                              # найдем коэффициент коррелящияя для интересующих нас столбцов
                   'city_сenters_nearest',                    # для этого создадим цикл с перебором по столбцам датафрейма
                   'rooms',
                   ]

for corr_value in columns_to_corr:
        print('Коэффициент корреляции для ', corr_value, 'равен' )
        print(df_good['last_price'].corr(df_good[corr_value]))
```

Мы расчитали коээфиценты Пирсона для каждого интересующего нас столбца. Он помогает определить наличие линейной связи между величинами и принимает значения от -1 до 1.
Если с ростом первой величины растёт вторая, то коэффициент корреляции Пирсона — положительный.
Если при изменении одной величины другая остаётся прежней, то коэффициент равен 0.
Если рост одной величины связан с уменьшением другой, коэффициент отрицательный.
Чем ближе коэффициент корреляции Пирсона к крайним значениям 1 или -1, тем теснее линейная взаимосвязь. Если значение близко к нулю, значит, линейная связь слабая либо отсутствует вовсе. Бывает, что коэффициент нулевой не потому, что связи между значениями нет, а из-за того, что у неё более сложный, не линейный характер. Потому-то коэффициент корреляции такую связь не берёт.

Исходя из этого мы можем говорить о более тесной связи стоимости квартиры с ее общей площадью, при этом чем больше квадратных метров, тем выше стоимость жилья - логично.

Коэффициет показывающий наличие связи стоимости жилья с близостью к центру города имеет отрицательное значение, следовательно чем меньше расстояние до центра города, тем выше стоимость. Связь между показателями не сильно выраженная.

Так же можно увидеть невыраженную связь между количеством комнат и стоимостью объекта. Чем больше комнат, тем больше стоимость.

Построим диаграммы распределения для этих величин.
